```python
import numpy as np
```

```python
from time import time
#Función para calcular el tiempo de ejecución
def calcular_tiempo(f):
    def wrapper(*args, **kwargs):        
        inicio = time()       
        resultado = f(*args, **kwargs)       
        tiempo = time() - inicio
        print("Tiempo de ejecución para algoritmo: "+str(tiempo))
        return resultado
    return wrapper
```

<b>Problema Ordenación Método Quick Sort</b>

```python
A = [9187, 244, 4054, 9222, 8373, 4993, 5265, 5470, 4519, 7182, 2035, 3506, 4337, 7580, 2554, 2824, 8357, 4447, 7379]
```

```python
def quick_sort(A):
    if (len(A)==1):
        return A
    elif (len(A)==2):
        return [min(A),max(A)]
    else:
        #pivote = (A[0]+A[1]+A[3])/3
        pivote = np.mean(A) #Complejidad n
        IZQ = []
        DER = []
        for x in A: #Complejidad 2*T(n/2)
            if (x<=pivote):
                IZQ.append(x)
            else:
                DER.append(x)
    return quick_sort(IZQ) + quick_sort(DER)
```

```python
@calcular_tiempo
def QS(A):
    return quick_sort(A)

print('Lista original:')
print(A)
print('\nLista ordenada:')
print(QS(A))
```

<b>Problema Cambio de Monedas</b>

```python
sm = [25,10,5,1]
```

```python
@calcular_tiempo
def cambio_monedas(C, sm):
    SOLUCION = [0 for i in range(len(sm))]
    VALOR_ACUMULADO = 0
    for i in range(len(sm)):
        monedas = int((C - VALOR_ACUMULADO)/sm[i])
        SOLUCION[i] = monedas
        VALOR_ACUMULADO += monedas*sm[i]
        
        if (C==VALOR_ACUMULADO):
            return SOLUCION
    print(SOLUCION)
    return

```

```python
cambio_monedas(99,sm)
```

<b>Problema de las Cuatro Reinas</b>

```python
def escribe(S):
    n = len(S)
    for x in range(n):
        print(" ")
        for i in range(n):
            if S[i] == x+1:
                print(" X ", end = " ")
            else:
                print(" - ", end = " ")
```

```python
N = 8
solucion = [0 for i in range(N)]
etapa = 0

def es_prometedora(SOLUCION, etapa):
    for i in range(etapa+1):
        if SOLUCION.count(SOLUCION[i])>1:
            return False
        for j in range(i+1, etapa+1):
            if (abs(i-j) == abs(SOLUCION[i]-SOLUCION[j])):
                return False
    return True 
    
def reinas(N, solucion, etapa):
    for i in range(1,N+1):
        solucion[etapa] = i
    
        if es_prometedora(solucion, etapa):
            if (etapa == N-1):
                print('\n\nLa solución es:')
                print(solucion)
                escribe(solucion)
            else:
                reinas(N, solucion, etapa+1)
        else:
            None
        
        solucion[etapa] = 0
```

<b>Problema de las Cuatro Reinas (Método iterativo)</b>

```python
def pinta(S):
    n = len(S)
    for x in range(n):
        print(" ")
        for i in range(n):
            if S[i] == x:
                print(" X ", end = " ")
            else:
                print(" - ", end = " ")

def validar(SOLUCION, stage):
    if(stage>0): #Solo valida a partir de la etapa 1
        for i in range(stage):
            if ((SOLUCION[i]==SOLUCION[stage]) or (abs(SOLUCION[i]-SOLUCION[stage])==(abs(i-stage)))):
                return False
        return True
    else: return True

def reinas_iterativo(num_reinas):
    etapa = 0
    solucion = [-1 for i in range(num_reinas)]
    while (etapa>=0):
        solucion[etapa] = solucion[etapa] + 1
        while ((solucion[etapa]<=num_reinas) and (not validar(solucion, etapa))):
            solucion[etapa] = solucion[etapa] + 1
        if (solucion[etapa]<=(num_reinas-1)):
            if (etapa == (num_reinas-1)):
                print('\n\nLa solución es:')
                print(solucion)
                pinta(solucion)
            else:
                etapa = etapa + 1
                solucion[etapa] = -1
        else:
            solucion[etapa] = -1
            etapa = etapa - 1


                
```

```python
reinas_iterativo(4)
```
