Diseña un algoritmo para resolver el problema por fuerza bruta

La resolución del algoritmo mediante fuerza bruta es dispendiosa porque requiere calcular todas las posibles 362.880 soluciones. Hay que calcular todos los posibles subconjuntos de 5 elementos a partir de la lista de 9 y para cada uno de ellos, calcular todas las posibles permutaciones (cambiando el orden dentro de cada subconjunto). Cada una de ellas, se debe poder combinar a su vez, con todas las posibles permutaciones del conjunto de operadores.<br>
Por lo tanto, la solución al problema se divide en varios algoritmos necesarios para la función generadora de soluciones. La primera función, <b>subconjuntos</b> permite obtener todos los posibles subconjuntos de 5 elementos a partir de una lista con los números de 1 al 9.<br>
Ya que el problema exige probar con los números en cualquier orden, se tiene la función <b>permutaciones</b> para calcular todas las posibles variantes (en cuanto a orden) de un conjunto determinado. Esta función se aplica tanto para los subconjuntos de números como para el conjunto de operadores.
Finalmente, hay una rutina principal que hace las combinaciones de cada permutación numérica con cada permutación de operadores en una única lista para aplicar la función de evaluación.

```python
#Algoritmo para obtener todos los posibles subconjuntos de 5 elementos a partir del 
#conjunto lista = ['1','2','3','4','5','6','7','8','9']

#Los resultados se van agregando a una lista de listas 
lista_subconjuntos = []

def subconjuntos(L, n, k, indice, temp, i): 
    if(indice == k): 
        #Se almacena cada nuevo número en una lista parcial
        parcial = []
        for j in range(k): 
            parcial.append(temp[j])
        #Cuando la lista parcial se ha recorrido completamente, se agrega a la lista de listas
        lista_subconjuntos.append(parcial)
        return
    else:
        #Mientras el contador i sea menor que la longitud de la lista de elementos original
        if(i < n): 
            #Calcula el siguiente elemento y lo agrega a la lista temporal
            temp[indice] = L[i]
            #Llama de manera recursiva la función para indice + 1 e indice
            subconjuntos(L, n, k, indice + 1, temp, i + 1) 
            subconjuntos(L, n, k, indice, temp, i + 1)
            return #resultado

lista = [str(i+1) for i in range(9)] 
  
k = 5
n = len(lista) 
datos = list(range(k))
subconjuntos(lista, n, k, 0, datos, 0)
print(lista_subconjuntos)
```

```python
#Se calcula el número de posibles subconjuntos de 5 elementos a partir de la lista de 9 elementos, para verificar
#si el cálculo teórico es correcto n!/(k!(n-k)!) =  9!/(5!(9-5)!) = 126.
print('La cantidad de subconjuntos de ' + str(k) + ' elementos a partir de un conjunto de ' + str (n) + ' elementos es:')
print(len(lista_subconjuntos))
```

<b>Algoritmo para obtener todas las posibles permutaciones de una lista</b><br>
Para cada uno de los subconjuntos, se debe probar con todas las posibles permutaciones.<br>
El siguiente algoritmo recibe por entrada una lista y entrega una lista de listas con todas las permutaciones.<br>
Este se usara tanto para cada uno de los subconjuntos del paso anterior, así como con la lista de operadores matermáticos.

```python
#Algoritmo para la obtención de todas las posibles permutaciones de una lista de n elementos
#Recibe una lista y devuelve una lista de listas
n = 5
lista = [str(i+1) for i in range(n)]

def permutaciones(L):
    if len(L)==2:
        resultado = [[L[0],L[1]],[L[1],L[0]]]
    else:
        resultado = []
        for i in range(len(L)):
            A = L[:i] + L[i+1:]
            perm = permutaciones(A) #Procedimiento recursivo en el que A es L menos un elemento diferente cada vez
            for j in range(len(perm)):
                res = [L[i]] + perm[j]
                resultado.append(res)
    return resultado
```

```python
#Se verifica que la cantidad de permutaciones coincida con el valor calculado teóricamente.
#Para un conjunto de n elementos, debe ser igual a n!. Por lo tanto, para cada subconjunto de 5 elementos, 
#hay 5! = 120 posibilidades y para los operadores, hay 4! = 24 posibilidades.
conjunto_prueba = [str(i+1) for i in range(5)]
operadores = ['+','-','*','/']

perm_conj_prueba = permutaciones(conjunto_prueba)
perm_conj_oper = permutaciones(operadores)

print('La cantidad de permutaciones del conjunto de prueba es: ')
print(len(perm_conj_prueba))
print('La cantidad de permutaciones del conjunto de operadores es: ')
print(len(perm_conj_oper))
```

Para calcular todas las posibles soluciones, el programa principal debe recorrer la variable <b>lista_subconjuntos</b> generada por la función <b>subconjuntos</b> y para cada uno de los elementos de la lista, que es a su vez una lista de 5 elementos, aplicar la función <b>permutaciones</b> que devuelve todas las posibles permutaciones para cada lista. Mediante un bucle adicional, se deben combinar cada permutación de números con cada una de las permutaciones de operadores, que se encuentran en la variable <b>perm_conj_oper</b>.
Esto genera 126x120x24 = 362.880 soluciones diferentes.
Aunque bastaría con detener el programa cuando una solución sea hallada, se van a calcular todas las soluciones para efectos de analizar el universo de soluciones en detalle y para calcular con exactitud el mayor y el menor resultado posibles. Hacer esto, también permite conocer las diferentes soluciones para un mismo valor, si es que las tiene.

<h2>Programa principal para resolución por fuerza bruta<h2>

```python
operadores = ['+','-','*','/']
perm_conj_oper = permutaciones(operadores)
#Se usa la variable lista_subconjuntos calculada en el primer algoritmo de este notebook, que contiene los 126 posibles
#subconjuntos de 5 elementos que se genera a partir de lista = ['1','2','3','4','5','6','7','8','9']

esperado = 2 #Este es el valor para el cual se van a buscar soluciones.
valor_minimo_posible = 0
valor_maximo_posible = 0

soluciones = []
valores = []
f = open("soluciones.csv","w+")

contador = 0

for x in lista_subconjuntos:
    permutaciones_x = permutaciones(x)
    for y in permutaciones_x:
        for z in perm_conj_oper:
            #y es una lista de 5 números
            #z es una lista de 4 operadores
            #Se crea una lista vacía de 9 elementos que se irá poblando con números y signos alternadamente
            expresion = ['' for j in range(9)]
            for i in range(5):
                pos = 2*i
                expresion[pos] = y[i]
                if (pos+1)<len(expresion):
                    expresion[pos+1] = z[i]
            cadena = ''.join(expresion)
            evaluacion = eval(cadena)
            #Se agregan todas las soluciones y los valores respectivos para su posterior análisis 
            soluciones.append(cadena)
            valores.append(evaluacion)
            linea = str(contador) + ';' + cadena + ";" + str(evaluacion) + '\n'
            #print(contador)
            f.write(linea)
            #Se cuentan las iteraciones
            contador += 1 
            if evaluacion > valor_maximo_posible:
                valor_maximo_posible = evaluacion
            if evaluacion < valor_minimo_posible:
                valor_minimo_posible = evaluacion
            
            if eval(cadena)==esperado:
                print('La solución es:' + cadena + ' = ' + str(esperado))

f.close
print('Valor máximo encontrado: ' + str(valor_maximo_posible))
print('Valor mínimo encontrado: ' + str(valor_minimo_posible))
print('Número de iteraciones: ' + str(contador))
```

Calcula la complejidad del algoritmo por fuerza bruta

El algoritmo tiene una complejidad definida por una función O(n!) debido a que tanto el cálculo de los posibles subconjuntos como las posibles permutaciones dependen del factorial del número de elementos.
Haciendo el análisis de OE de la función <b>permutaciones</b>, se llega a la conclusión que es una función de la forma T(n) = n^3 + 3n^2 + n (1 + T(n-1)) + 3. Como se puede ver, queda un término de la forma nT(n-1) y la función que es recurrente, se llama a si misma al menos 4 veces. Con cada llamada este n se va multiplicando por n-1, n-2, es decir, n!.

<h1>Resolución del problema mediante un algoritmo heurístico</h1><br>
A diferencia del método de fuerza bruta, es posible hallar una solución para varios casos en una fracción del tiempo

```python
import random

def hallar_solucion(esperado, limite):
    #El primer argumento es el valor para el que se debe hallar solución
    #El segundo argumento es el número máximo de iteraciones para buscar solución
    conj = ['1','2','3','4','5','6','7','8','9']
    oper = ['+','-','*','/']

    max = 0
    min = 0

    if (esperado > 78.83333333333333) or (esperado < -70.71428571428571):
        print('El valor se encuentra por fuera del rango posible [-70.71428571428571, 78.33333333333333]')
    elif not(esperado==int(esperado)):
        print('El valor debe ser entero')
    else: 
        resultado = esperado + 1
        contador = 0
        while (not(resultado == esperado)) and (contador<limite):
            random.shuffle(conj)
            random.shuffle(oper)
            expresion = ['' for x in range(9)]
            for i in range(5):
                pos = 2*i
                expresion[pos] = conj[i]
                if (pos+1)<len(expresion):
                    expresion[pos+1] = oper[i]
            cadena = ''.join(expresion)
            resultado = eval(cadena)
            if resultado > max:
                max = resultado
            if resultado < min:
                min = resultado
            contador += 1
        if eval(cadena)==esperado:
            print('La solución es:' + cadena + ' = ' + str(esperado))
        else:
            print('No se encontró solución después de ' + str(limite) + ' iteraciones.')
        print('Valor máximo encontrado: ' + str(max))
        print('Valor mínimo encontrado: ' + str(min))
        print('Número de iteraciones: ' + str(contador))
        
valor = 5
hallar_solucion(valor, 10000)
```

Como se puede observar en el ejemplo anterior, en algunos casos el algoritmo es capaz de encontrar soluciones muy rápidamente, aunque esto obviamente es completamente aleatorio. En el peor de los casos, es posible que el algoritmo no encuentre una solución dentro del número de iteraciones, pero este número no depende directamente del tamaño de los conjuntos de números u operadores. Depende del valor máximo de iteraciones que se haya definido. Considerando que las 362.880 posibles soluciones se concentran en 3533 valores distintos, y de las 362.880 posibles soluciones solo 90.000 arrojan valores enteros que se pueden mapear en 147 números enteros distintos, calculamos el número mínimo de iteraciones para tener una probabilidad específica de encontrar una solución a partir de un valor dado, así:

```python
from math import log
print('Probabilidad\tIteraciones i>\n')
for i in range(5):
    p = 0.5 + i/10
    print(str(p) + '\t\t' + str(log(1-p)/log(1-1/362280)))
```

En el análisis que se hace en el archivo soluciones.xlsx se encontró que entre los 3533 posibles valores que toma la evaluación de la función, sólo 147 son números enteros, pero por ejemplo, el 5 se puede representar de 2880 maneras diferentes, mientras que el -69 sólo tiene 16 posibles soluciones (y es entre los enteros, el que menos soluciones tiene). Esto indica que la probabilidad de cada valor es distinta y el peor caso sería el -69, cuya probabilidad es de 16/362880

Dado que el algoritmo incorpora la función shuffle de Python, es necesario ver su implementación, la cual se detalla a continuación:

```python
def shuffle(self, x, random=None, int=int):
    """x, random=random.random -> shuffle list x in place; return None.
    Optional arg random is a 0-argument function returning a random
    float in [0.0, 1.0); by default, the standard random.random.
    """
    if random is None:
        random = self.random
        for i in reversed(xrange(1, len(x))):
            # pick an element in x[:i+1] with which to exchange x[i]
            j = int(random() * (i+1))
            x[i], x[j] = x[j], x[i]
```

Esta función es de orden <b>n</b>. Llama internamente a la función <b>reversed</b>, que simplemente invierte el orden de iteración del rango dado. Por lo tanto, La máxima complejidad del algoritmo depende la cantidad de iteraciones, multiplicada por n más un número de operaciones elementales.
La cantidad de operaciones es una función del tipo i*(an + c), luego es de O(n), donde i es el número de iteraciones, a es la cantidad de veces que se corre un bucle simple y c un número de operaciones elementales. También se usa la función <b>len</b>, la cual se estima que es de O(n) ya que debe recorrer la lista una vez para obtener la cantidad de elementos de esta.
